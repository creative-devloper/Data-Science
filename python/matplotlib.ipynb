### SCATTER CHART

- bivariate analysis
- num vs num

```python
batter = pd.read_csv('batter.csv')
```

```python
temp_df = batter.sample(100).head(25)
```

```python
temp_df
```

```python
plt.figure(figsize=(10,7))
plt.scatter(temp_df['avg'],temp_df['strike_rate'],marker='+') #s = temp_df['runs']
plt.axhline(85,color="#C0AD3F")
plt.axvline(22,color="#DA49B3")
for i in range(temp_df.shape[0]):
    plt.text(temp_df['avg'].values[i],temp_df['strike_rate'].values[i],temp_df['batter'].values[i])
plt.show()
```

```python
batsman
```

```python
plt.barh(batsman['batsman'],batsman['2015'],color='purple')
```

```python
plt.bar(batsman['batsman'],batsman['2015'],label='2015')
plt.bar(batsman['batsman'],batsman['2016'],bottom=batsman['2015'],label='2016')
plt.bar(batsman['batsman'],batsman['2017'],bottom=(batsman['2015']+batsman['2016']),label='2017')
plt.legend()
```

```python
plt.bar(np.arange(batsman.shape[0]),batsman['2016'],width=0.3,label='2016')
plt.bar(np.arange(batsman.shape[0])-0.3,batsman['2015'],width=0.3,label='2015')
plt.bar(np.arange(batsman.shape[0])+0.3,batsman['2017'],width=0.3,label='2017')
plt.legend()
plt.xticks(np.arange(batsman.shape[0]),batsman['batsman'])
plt.show()
```

```python
# plt.figure(figsize=(10,10))
fig,ax = plt.subplots(nrows=2,ncols=1,sharex=True)
fig.set_figheight(7)
fig.set_figwidth(7)
ax[0].scatter(batter['avg'],batter['strike_rate'],c='red')
ax[1].scatter(batter['avg'],batter['runs'])

ax[0].set_xlabel('Avg')
ax[0].set_ylabel('Strike Rate')
ax[1].set_xlabel('Avg')
ax[1].set_ylabel('Runs')

ax[0].set_title("Avg vs SR")
ax[1].set_title("Avg vs Runs")

```

```python
# plt.figure(figsize=(10,10))
fig,ax = plt.subplots(nrows=2,ncols=2,sharex=True)
fig.set_figheight(5)
fig.set_figwidth(12)
ax[0,0].scatter(batter['avg'],batter['strike_rate'],c='red')
ax[1,0].scatter(batter['avg'],batter['runs'])

ax[0].set_xlabel('Avg')
ax[0].set_ylabel('Strike Rate')
ax[1].set_xlabel('Avg')
ax[1].set_ylabel('Runs')

ax[0].set_title("Avg vs SR")
ax[1].set_title("Avg vs Runs")

```

```python
fig = plt.figure(figsize=(10,7))

ax1 = fig.add_subplot(2,2,1)
ax1.scatter(batter['avg'],batter['strike_rate'])

ax2 = fig.add_subplot(2,2,2)
ax2.scatter(batter['avg'],batter['runs'])


ax3 = fig.add_subplot(2,2,3)
ax3.hist(batter['strike_rate'])


ax4 = fig.add_subplot(2,2,4)
ax4.hist(batter['runs'])
```

```python
fig = plt.figure(figsize=(10,7))
ax1 = fig.add_subplot(projection='3d')
ax1.scatter3D(batter['avg'],batter['strike_rate'],batter['runs'])
ax1.set_xlabel("AVG")
ax1.set_ylabel("SR")
ax1.set_zlabel("runs")
```
