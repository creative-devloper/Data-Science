two types of fucntions

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

#### relational plot (relplot)

it is used to showcase the relationship between different variables,
- lineplot
- scatterplot

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.scatterplot(data=tips, x='total_bill',y='tip',hue='sex',style='time',size='size')
```

```python
sns.relplot(data=tips, x='total_bill',y='tip',hue='sex',style='time',size='size',kind='scatter')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',col='sex',row='day')
```

```python
df = px.data.gapminder()
```

```python
df
```

```python
India = df[df['country']=='India']
```

```python
India
```

```python
sns.lineplot(data=India,x='year',y='lifeExp')
```

```python
sns.relplot(data=India,x='year',y='lifeExp',kind='line')
```

```python
temp_df = df[df['country'].isin(['Germany','India','Brazil'])]
```

```python
temp_df
```

```python
sns.relplot(data=temp_df, x= 'year',y='lifeExp',kind='line',hue='country',style='continent')
```

```python
pivot = df.pivot_table(index='country',columns='year',values='lifeExp')
```

```python
Europe = df[df['continent']=='Europe'].pivot_table(index='country',columns='year',values='lifeExp')
```

```python
plt.figure(figsize=(20,15))
sns.heatmap(Europe,annot=True,linewidths=0.2)#,cmap='viridis'
```
